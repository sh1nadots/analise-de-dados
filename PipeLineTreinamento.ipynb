#Autenticação e paginação

```python
#paginação: transitar entre as páginas, extrair informações de cada página da API
api_base_url = 'http://api.github.com'
owner = 'amzn' #username amazon
url = f'{api_base_url}/users/{owner}/repos'
url
```

```python
repos_list= []
for page_num in range(1,6):
  try:
    url_page = f'{url}?page={page_num}'
    response = requests.get(url_page, headers=headers)
    repos_list.append(response.json())
  except:
    repos_list.append(None)
```

```python
len(repos_list)
```

```python
len(repos_list[0])
```

#aplicando transformações para limpeza de informações

```python
repos_list
```

```python
repos_list[0][2]['name']
```

```python
repos_name = []
for page in repos_list:
    for repo in page:
        repos_name.append(repo['name'])
```

##Linguagens utilizadas em todos repositórios

```python
repos_list[1][1]['language']
```

```python
repos_language = []
for page in repos_list:
  for repo in page:
    repos_language.append(repo['language'])
```

##formato de arquivo

```python
import base64
```

```python
with open('amazon.csv', 'br') as file:
  file_content = file.read()

encoded_content = base64.b64encode(file_content)
```

#Upload de arquivo com PUT

```python
api_base_url = 'https://api.github.com'
username = 'millenagena'
repo = 'linguagens-utilizadas'
path = 'amazon.csv'

url = f'{api_base_url}/repos/{username}/{repo}/contents/{path}'
url
```

```python
data = {
    'message': 'Adicionando um novo arquivo',
    'content': encoded_content.decode('utf-8')
}

response = requests.put(url, json=data, headers=headers)
response.status_code
```
