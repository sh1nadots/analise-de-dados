```python
valores_exames_v3 = valores_exames_v2.drop(columns=["exame_29", "exame_4"])

def classificar(valores):
    SEED = 1234
    random.seed(SEED)
    x_train, x_test, y_train, y_test = train_test_split(valores, diagnostico,test_size = 0.5)

    classificador = RandomForestClassifier(n_estimators = 100)
    classificador.fit(x_train, y_train)
    print("Resultado da classificaÃ§Ã£o %.2f%%" %(classificador.score(x_test, y_test)*100))

classificar(valores_exames_v3)
```

```python
valores_exames_v6 = valores_exames_v1.drop(columns=(["exame_4", "exame_29", "exame_3", "exame_24"]))
```

```python
SEED= 1234
random.seed(SEED)

x_train, x_test, y_train, y_test = train_test_split(valores_exames_v6, diagnostico, test_size = 0.3)

selecionar_kmelhores.fit(x_test, y_test)
treino_kbest = selecionar_kmelhores.transform(x_train)
teste_kbest = selecionar_kmelhores.transform(x_test)
teste_kbest.shape
```

```python
classificador = RandomForestClassifier(n_estimators=100, random_state=1234)
classificador.fit(treino_kbest, y_train)
print("Resultado da classificaÃ§Ã£o %.2f%%" %(classificador.score(teste_kbest,y_test)*100))
```

```python
from sklearn.feature_selection import RFE

SEED= 1234
random.seed(SEED)

x_train, x_test, y_train, y_test = train_test_split(valores_exames_v6, diagnostico,test_size = 0.3)

classificador = RandomForestClassifier(n_estimators=100, random_state=1234)
classificador.fit(x_train, y_train)
selecionador_rfe = RFE(estimator = classificador, n_features_to_select = 5, step = 1)
selecionador_rfe.fit(x_train, y_train)
rfe_train = selecionador_rfe.transform(x_train)
rfe_test = selecionador_rfe.transform(x_test)
classificador.fit(rfe_train, y_train)

matriz_confusao = confusion_matrix(y_test,classificador.predict(rfe_test))
plt.figure(figsize = (10, 8))
sns.set(font_scale= 2)
sns.heatmap(matriz_confusao, annot = True, fmt = "d").set(xlabel = "PrediÃ§Ã£o", ylabel= "Real")

print("Resultado da classificaÃ§Ã£o %.2f%%" %(classificador.score(rfe_test,y_test)*100))
```

```python
from sklearn.feature_selection import RFECV

SEED = 1234
random.seed(SEED)

x_train, x_test, y_train, y_test = train_test_split(valores_exames_v6, diagnostico, test_size = 0.3)

classificador = RandomForestClassifier(n_estimators=100, random_state=1234)
classificador.fit(x_train, y_train)
selecionador_rfecv = RFECV(estimator = classificador, cv = 5, step = 1, scoring ='accuracy')
selecionador_rfecv.fit(x_train, y_train)
rfecv_train = selecionador_rfecv.transform(x_train)
rfecv_test = selecionador_rfecv.transform(x_test)
classificador.fit(rfecv_train, y_train)

matriz_confusao = confusion_matrix(y_test,classificador.predict(rfecv_test))
plt.figure(figsize=(10,8))
sns.set(font_scale=2)
sns.heatmap(matriz_confusao, annot = True, fmt = 'd').set(xlabel= 'predict', ylabel = 'real')

print('Resultado da classificacao foi de %.2f%%' %(classificador.score(rfecv_test, y_test)*100))
```
