Missão1: time de vendas

precisamos construir uma visualização para indicar os 10 maiores clientes por total de venda.

com isso, a equipe de vendas pode pensar em estratégias para mantes esses clientes comprando em nossa loja

```
df_cliente_vendas = dados.groupby(['nome_cliente'])['vendas'].sum().nlargest(10).copy()
df_cliente_vendas = df_cliente_vendas.reset_index()
df_cliente_vendas
```

```
df_cliente_vendas.columns = ['Clientes','Vendas']
df_cliente_vendas['Ranking'] = df_cliente_vendas.index + 1
df_cliente_vendas.set_index('Ranking', inplace = True)
```

```
df_cliente_vendas
```

```
s.format({'Vendas':'R$ {:,.2f}'})
s
```

#Missão 2: Time comercial

Precisamos construir uma visualização que permita associar o total ganho em vendas e o total lucrado por tipo de produto.

Isso vai permitir que o time comercial possa focar em estratégias para alavancar as rendas que ganham com cada tipo de produto

```
df_venda_lucro = dados.groupby(['tipo_produto'])[['vendas', 'lucro']].sum()
df_venda_lucro.index.name = 'tipo produto'
df_venda_lucro
```

```
estilo_produto = df_venda_lucro.style
estilo_produto.format('R$ {:,.2f}').highlight_max(color='lightgreen').highlight_min(color='#E6BAB7')
estilo_produto
```

```
estilo_produto.format('R$ {:,.2f}').background_gradient(cmap='Greens')
estilo_produto
```

```
cabecalho = {
    'selector':'th', #selecionando e definindo o cabeçalho
    'props':'font-wight: bold; font-family: Arial; text-align: center; text-transform: capitalize;'  #edições de texto

}
estilo_produto.set_table_styles([cabecalho], overwrite=False)
```

#Missão 3: Time de Logística

Precisamos construir uma visualização que permita entender a distribuição de pedidos por região do Brasil.
Isso vai permitir que o time de logística entenda onde é mais interessante focar o material e pessoal

```
df_regiao = pd.DataFrame(dados['regiao'].value_counts())
df_regiao.columns = ['Nº pedidos']
df_regiao.index.name = 'Região'
df_regiao
```

```
import numpy as np
```

```
porcentagem = df_regiao['Nº pedidos'].to_numpy()
porcentagem = 100 * porcentagem/porcentagem.sum()
porcentagem
```

```
df_regiao['Porcentagem Região'] = porcentagem
```

```
df_regiao
```

```
estilo_regiao = df_regiao.style
estilo_regiao
```

```
cabecalho = {
    'selector': 'th',
    'props': 'font-weight: bold; font-family: Arial; text-align: right; background-color:lightblue;'
}
celulas = {
    'selector':'td',
    'props': 'background-color:white;'
}
estilo_regiao.set_table_styles([cabecalho, celulas])
```

```
estilo_regiao.format({'Porcentagem Região':'{:.2f} %'}).bar(subset = 'Porcentagem Região', vmin=0, vmax=100.0, color= '#9CD33B')
```

#Missão 4: Time de Logística

Novamente o time de logística deseja repensar sua estrutura e organização do fornecimento de produtos e para isso, foi pensada em uma visualização que permita identificar padrão de qualidade de produtos solicitados por mês por departamento.

```
df_meses = dados.copy()
df_meses = df_meses.sort_values('data_pedido')
df_meses
```

```
df_meses['meses']= df_meses['data_pedido'].dt.strftime('%Y - %b')
df_meses = df_meses.reset_index(drop=True)
df_meses
```

```
vendas_mensais = df_meses.pivot_table(index = 'departamento', columns = 'meses', values = 'quantidade', aggfunc = 'sum', sort = False)
vendas_mensais
```

```
estilo_mensal = vendas_mensais.style
estilo_mensal
```

```
estilo_mensal.set_sticky(axis = 'index')
```

```
colunas = {
    'selector': '.col_heading',
    'props': 'font-weight: bold; font-family: Arial; background-color:lightblue;'
}
tabelas = {
    'selector': 'td,th',
    'props': 'font-weight: normal; font-family: Arial; text-align: left;'
}
index = {
    'selector': '.index_name',
    'props': 'font-weight: bold; font-family: Arial; text-align: right;'
}

estilo_mensal.set_table_styles([colunas, tabelas, index], overwrite=False)
```

#Missão 5: Relatório de performance

O relatório de performance é um documento muito importante para a empresa.
Nele, todos que trabalham na loja podem entender como anda a empresa.
Por conta disso, tabelas bem construídas são essenciais para mostrar os dados da melhor forma.

Portanto, precisamos contruir uma visualização que permita entender a relação entre tipos de clientes e modo de envio de produtos de acordo com as vendas.
Com isso, será possivel destacar no relatório qual o modo de envio e quais tipos de modo de envio mais geram ganho em vendas.

```
df_cliente = dados.pivot_table(index = 'segmento_cliente', columns = 'modo_envio', values = 'vendas', aggfunc = 'sum')
df_cliente
```

```
df_cliente['Total'] = df_cliente.sum(axis=1)
df_cliente.loc['Total'] = df_cliente.sum()
df_cliente
```

```
compra_cliente = df_cliente.style.format('{:,.2f}')
compra_cliente
```

```
tabela = {
    'selector':'td,th:not(.index_name)',
    'props':'font-weight: normal; font-family: Arial; text-align: center; background-color: white; color:black;'
}
index = {
    'selector':'index_name',
    'props':'font-weight: normal; font-family: Arial; text-align: right; font-style: italic; color:#696969;'
}
compra_cliente.set_table_styles([tabela,index])
```

```
compra_cliente.set_table_styles({
    'Total': [{
        'selector':'th',
        'props':'border-top: 1px solid #181818'
    },
              {
        'selector':'td',
        'props':'border-top: 1px solid #181818'
    }],
          'B2B':[{
        'selector':'th',
        'props':'border-top: 1px solid #181818'
    },
    {
        'selector':'td',
        'props':'border-top: 1px solid #181818'
    }]
}, overwrite = False, axis=1)
```

```
compra_cliente.set_table_styles({
    'Total': [{
        'selector':'.true',
        'props':'background-color: #D8D8D8'
    }]
}, overwrite = False, axis=0)

cores_coluna = pd.DataFrame(['false','true', 'false'], index= df_cliente['Total'].index, columns= ['Total'])

compra_cliente.set_td_classes(cores_coluna)
```

```
compra_cliente.set_table_styles({
    'Total': [{
        'selector':'.true',
        'props':'background-color: #D8D8D8'
    }]
}, overwrite = False, axis=0)

cores_total = pd.DataFrame(['false','false', 'true'], index= df_cliente['Total'].index, columns= ['Total'])

compra_cliente.set_td_classes(cores_total)
```
